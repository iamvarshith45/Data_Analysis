```python
# assign weights to a subset of stocks (example: RELIANCE.NS, HDFCBANK.NS, ICICIBANK.NS) to check portfolio analysis

weights = [0.4, 0.35, 0.25]
portfolio_data = nifty50_data[['RELIANCE.NS', 'HDFCBANK.NS', 'ICICIBANK.NS']]

# calculate daily returns
daily_returns = portfolio_data.pct_change().dropna()

# calculate portfolio returns
portfolio_returns = (daily_returns * weights).sum(axis=1)

# display portfolio returns
portfolio_returns.head()
```

pct_change() computes daily percentage returns for each stock -
 Daily_Return = (Current_Price - Previous_Price) // Previous_Price and
 dropna() removes the first row (which will have NaN since there's no previous price to calculate change).

through this process we can analyze overall performance of a portfolio rather than individual stocks, each value represents the percentage change in the portfolios value for a particular day. For example, a return of -0.002790 on the first day indicates a 0.279% decrease in the portfolios value,
while 0.004495 on the second day indicates a 0.4495% increase.
These values help in tracking the portfolios daily performance over time and manage expenses and mitigate risk.


```python
volatility = daily_returns.std()
risk_metrics = pd.DataFrame({'Volatility (Std Dev)': volatility})
risk_metrics
```

RELIANCE has a volatility of 0.87%, HDFCBANK has 0.69%, and
ICICIBANK has 1.16%. This shows that ICICIBANK has the highest price fluctuations, while HDFCBANK is the least price fluctuations with respect to market conditions.

```python
import seaborn as sns

# Set Seaborn style
sns.set_style("whitegrid")

# Calculate moving averages for RELIANCE
nifty50_data['RELIANCE_5d_MA'] = nifty50_data['RELIANCE.NS'].rolling(window=5).mean()
nifty50_data['RELIANCE_20d_MA'] = nifty50_data['RELIANCE.NS'].rolling(window=20).mean()

# Create a figure
plt.figure(figsize=(12, 6))

# Plot RELIANCE stock price
sns.lineplot(x=nifty50_data['Date'], y=nifty50_data['RELIANCE.NS'], label="RELIANCE.NS Price", color='blue')

# Plot 5-day moving average
sns.lineplot(x=nifty50_data['Date'], y=nifty50_data['RELIANCE_5d_MA'], label="5-Day MA", color='orange')

# Plot 20-day moving average
sns.lineplot(x=nifty50_data['Date'], y=nifty50_data['RELIANCE_20d_MA'], label="20-Day MA", color='green')

# plot
plt.title("Moving Averages for RELIANCE.NS", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Legend")

plt.show()
```

Sharpe Ratio is a measure of risk adjusted return that indicates how much return an investment generates for each unit of risk taken, and Helps compare multiple investment options, rather than just absolute returns.

Computed the ratio using the formula
(Mean Returns - Risk-Free Rate) // Volatility

A higher Sharpe Ratio means better risk-adjusted returns, simillarly low Sharpe Ratio indicates poor returns subject to risk.

```python

# average returns and volatility(Standard deviation)
mean_returns = daily_returns.mean()
volatility = daily_returns.std()

# assume a risk-free rate (yearly, divide by 252 trading days)
risk_free_rate = 0.04 / 252

# calculating Sharpe Ratio
sharpe_ratios = (mean_returns - risk_free_rate) / volatility

# neq Dataframe
table_data = pd.DataFrame({
    'Stock': sharpe_ratios.index,
    'Sharpe Ratio': sharpe_ratios.values.round(2)
})

# sorting the values
table_data = table_data.sort_values(by="Sharpe Ratio", ascending=False)

# Sharpe Ratios plot using Matplotlib
plt.figure(figsize=(10, 5))
plt.bar(table_data['Stock'], table_data['Sharpe Ratio'], color='skyblue')

plt.xlabel("Stock", fontsize=12)
plt.ylabel("Sharpe Ratio", fontsize=12)
plt.title("Sharpe Ratios for Selected Stocks", fontsize=14)

plt.show()
print(table_data)
```

RELIANCE.NS has a negative Sharpe Ratio (-0.05) means that RELIANCE.NS has underperformed hence stocks returns do not provide enough returns for the risk taken.

 HDFCBANK.NS has positive and low Sharpe Ratio (0.37) menas that the excess return is moderate when adjusted for risk.

ICICIBANK.NS has the highest risk adjusted return among the three stocks, showing that investors received a better returns based their risk tolerance. it is not the highest, but this Sharpe Ratio shows that this stock had a better  risk return tradeoff than other stocks.
